```python
import os
import pandas as pd

file_path = "jobs.xlsx"

if not os.path.exists(file_path):
    df = pd.DataFrame(columns=[
    "Job_ID",
    "Date_Accessed",
    "Company",
    "Job_URL",
    "Job_Title",
    "Keywords",
    "JD_Summary",
    "Job_Type",
    "Link_Status",
    "Content_Status",
    "Fetch_Note",
    "Status",
    "Applied_Date",
    "Next_Action",
    "Next_Action_Date",
    "CV_Tex_Path",
    "CV_PDF_Path",
    "Notes",
    "Last_Update"
    ])
    df.to_excel(file_path, index=False)
    print("jobs.xlsx 创建成功 ✅")
else:
    print("jobs.xlsx 已存在，未覆盖 ❗")
```
